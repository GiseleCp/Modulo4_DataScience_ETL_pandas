# Manipulando os dados

## Criando colunas numéricas

```python
# recuperar novamente arquivo original para criar novas colunas
url = 'https://raw.githubusercontent.com/GiseleCp/Modulo4_DataScience_ETL_pandas/refs/heads/main/aluguel.csv'
dados = pd.read_csv(url, sep=';')
dados.head()
```

```python
# criando uma coluna de gasto por mes
dados['Valor_por_mes'] = dados['Valor'] + dados['Condominio']
dados.head()
```

```python
# criando uma coluna com gasto anual
dados['Valor_por_ano'] = (dados['Valor_por_mes'] * 12) + dados['IPTU']
dados.head()
```

## Criando colunas categóricas

```python
# criando uma coluna de descricao
dados['Descricao'] = dados['Tipo'] + dados['Bairro']
dados.head()
```

```python
# criando uma coluna de descricao melhorando a separacao dos campos
dados['Descricao'] = dados['Tipo'] + 'em' + dados['Bairro']
dados.head()
```

```python
# criando uma coluna de descricao melhorando a separacao dos campos
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + dados['Quartos'].astype(str) + ' quartos ' + 'e' + dados['Vagas'].astype(str) + ' vagas '
dados.head()
```

```python
# criando uma coluna se possui suite, caso
dados['Possui_suite'] = dados['Suites'].apply(lambda x: 'Sim' if x > 0 else 'Não')
dados.head()
```

```python
dados.to_csv('dados_completos_dev.csv', index=False, sep=';')
```
